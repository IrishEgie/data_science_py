```
# Last 3 values
last_3_values = a[-3:]
print("Last 3 values:", last_3_values)

# 4th, 5th, and 6th values
subset_4_5_6 = a[3:6]
print("4th, 5th, and 6th values:", subset_4_5_6)

# All values except the first 12
subset_excluding_first_12 = a[12:]
print("Values excluding first 12:", subset_excluding_first_12)

# Only even numbers
even_numbers = a[a % 2 == 0]
print("Even numbers:", even_numbers)
```

# Broadcasting and Scalars


```
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
# Load the image
img = misc.face()

# Display the image
plt.imshow(img)
plt.axis('off')  # Turn off the axis
plt.show()
```

```
# Check the data type of img
data_type = img.dtype
print("Data type of img:", data_type)

# Check the shape of img
shape = img.shape
print("Shape of img:", shape)

# Check the number of dimensions
num_dimensions = img.ndim
print("Number of dimensions:", num_dimensions)

# Calculate the resolution of the image
resolution = shape[0] * shape[1]
print("Resolution of the image (width x height):", resolution)
```

```
# Convert to grayscale by averaging the RGB channels
gray_img = np.mean(img, axis=2).astype(np.uint8)

# Flip the grayscale image upside down
flipped_gray_img = np.flipud(gray_img)

# Display the flipped grayscale image
plt.imshow(flipped_gray_img, cmap='gray')
plt.axis('off')
plt.title("Flipped Grayscale Image")
plt.show()
```

```
# Rotate the color image 90 degrees clockwise
rotated_color_img = np.rot90(img, k=-1)  # k=-1 for 90 degrees clockwise

# Display the rotated color image
plt.imshow(rotated_color_img)
plt.axis('off')
plt.title("Rotated Color Image")
plt.show()
```

```
# Invert the color image (solarize)
inverted_color_img = 255 - img

# Display the inverted color image
plt.imshow(inverted_color_img)
plt.axis('off')
plt.title("Inverted Color Image")
plt.show()
```
